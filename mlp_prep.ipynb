```python
# Returns + Volatility

import numpy as np

prices = [100, 102, 101, 105, 110]

daily_returns = []
for prev,curr in zip(prices, prices[1:]):
    daily_returns.append((curr-prev)/prev)

expected_return = sum(daily_returns) / len(daily_returns)

variance = 0
for ret in daily_returns:
    variance += (ret - expected_return)**2
variance /= len(daily_returns)

volatility = np.sqrt(variance) * np.sqrt(252)

print(daily_returns)
print(volatility)
```

```python
# Rolling Average

import pandas as pd

arr = [1, 3, 2, 5, 4, 6, 7]
k = 3

df = pd.DataFrame({'vals': arr})
df['rolling_avg'] = df['vals'].rolling(k).mean()

print(df['rolling_avg'])
```

```python
# Top K Frequent

import heapq

nums = [1,1,1,2,2,3]
k = 2

counts = {}
for n in nums:
    counts[n] = counts.get(n,0) + 1

pq = []
for key,freq in counts.items():
    heapq.heappush(pq, [freq,key])
    if len(pq) > k:
        heapq.heappop(pq)

res = []
while pq:
    freq,key = heapq.heappop(pq)
    res.append(key)

print(res)
```

```python
# Mean Reversion Signal

prices = [100, 102, 101, 99, 98, 97, 98]
window = 3

df = pd.DataFrame({'prices': prices})
df['rolling_avg'] = df['prices'].rolling(window).mean()
df['signal'] = np.where(df['prices'] < df['rolling_avg'], 1,
               np.where(df['prices'] > df['rolling_avg'], -1, 0))

df.head(10)
```

```python
# Num Islands

grid = [
    ["1","1","0"],
    ["0","1","0"],
    ["1","0","1"]
]

def dfs(grid, row, col):
    if row < 0 or row >= len(grid) or col < 0 or col >= len(grid[row]) or grid[row][col] == '0':
        return
    
    if grid[row][col] == '1':
        grid[row][col] = '0'
    
    dfs(grid, row-1, col)
    dfs(grid, row+1, col)
    dfs(grid, row, col-1)
    dfs(grid, row, col+1)

islands = 0
for row,i in enumerate(grid):
    for col,j in enumerate(i):
        if j == '1':
            dfs(grid, row, col)
            islands += 1

print(islands)
```

```python
# Largest Drawdown

prices = [100, 120, 110, 130, 90, 95]

running_max = 0
drawdown = 0
for p in prices:
    if p > running_max:
        running_max = p
    
    dd = running_max - p
    if dd > drawdown:
        drawdown = dd

print(drawdown)
```
